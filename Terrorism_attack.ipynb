# Model Training and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report
```

models = {
   "LR": LogisticRegression(max_iter = 5000,solver='sag'),
    "RF": RandomForestClassifier(n_estimators=100),
    "DT" : DecisionTreeClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "XGB": XGBClassifier(n_estimators=100)
}

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=10000),
    'Random Forest': RandomForestClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier()
}
```

```python
for name,model in models.items():
    print(f'Training Model {name} \n--------------------------------------------------')
    model.fit(x_train_scaled,y_train)
    y_pred = model.predict(x_test_scaled)
    print(f'Training Accuracy: {model.score(x_train_scaled, y_train)}')
    print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
    print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
    print('-'*70)
```

## Hyperparametre Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grids = {
    'Logistic Regression': {'C': [0.001, 0.01, 0.1, 1, 10, 100]},
    'Random Forest': {'n_estimators': [50, 100, 200], 'max_depth': [None,10, 20, 30]},
    'Decision Tree': {'max_depth': [None, 10, 20, 30]},
    'Gradient Boosting': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]},
    'XGBoost': {'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 7], 'n_estimators': [100, 200, 300]}
}

# Perform hyperparameter tuning for each model
best_models = {}
scores = []
for name, model in models.items():
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids[name],
        scoring='accuracy',  # Choose an appropriate metric
        cv=5,  # Number of cross-validation folds
        verbose=2,
        n_jobs=-1,  # Use all available CPU cores
    )
    
    grid_search.fit(x_train_scaled, y_train)
    best_models[name] = grid_search.best_estimator_
    scores.append({
        'model': name,
        'best_score': grid_search.best_score_,
         'best_params': grid_search.best_params_
    })
```

```python
best_models_df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
best_models_df
```

```python
# Evaluate and compare models on the test set
results = {}
for name, model in best_models.items():
    y_pred = model.predict(x_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

# Print the results
for name, accuracy in results.items():
    print(f"{name} : {accuracy:.4f}")
```
