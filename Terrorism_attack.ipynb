```python
# Checking Null values
df.isnull().sum()
```

```python
#Return Percentage of missing values in each column
for column in df.columns:
    prct_missing = np.mean(df[column].isnull())
    print(f'{column} - {round(prct_missing*100)}%')
```

```python
# Drop columns that contain 25% or more missing values
for column in df.columns:
    prct_missing = np.mean(df[column].isnull())
    prct_missing = round(prct_missing*100)
    if prct_missing > 25 :
        df.drop([column],axis=1,inplace=True)
        print(f'Droping {column} Column as it contains - {prct_missing}% null values')
```

# Modelling and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,classification_report
```

```python
models = {
   "LR": LogisticRegression(max_iter = 5000,solver='sag'),
    "RF": RandomForestClassifier(n_estimators=100),
    "DT" : DecisionTreeClassifier(),
    "GradientBoosting": GradientBoostingClassifier(),
    "XGB": XGBClassifier(n_estimators=100)
}
```

```python
for name,model in models.items():
    print(f'Training Model {name} \n--------------------------------------------------')
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(f'Training Accuracy: {model.score(x_train, y_train)}')
    print(f'Testing Accuracy: {accuracy_score(y_test, y_pred)}')
    print(f'Testing Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}')
    print('-'*70)
```
