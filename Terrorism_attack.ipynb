## Analysis and Visualization

```python
# Checking Null values
df.isnull().sum()
```

```python
#Return Percentage of missing values in each column
for column in df.columns:
    prct_missing = np.mean(df[column].isnull())
    print(f'{column} - {round(prct_missing*100)}%')
```

```python
# Drop columns that contain 25% or more missing values
for column in df.columns:
    prct_missing = np.mean(df[column].isnull())
    prct_missing = round(prct_missing*100)
    if prct_missing > 25 :
        df.drop([column],axis=1,inplace=True)
        print(f'Droping {column} Column as it contains - {prct_missing}% null values')
```

## Terrorist Incident and Its Success

```python
sns.countplot(x= df['success'],order=df['success'].value_counts().index)
plt.title('Successed and not Successed Terrorist Incidents',size=20)
plt.show()
```

```python
plt.pie(df['success'].value_counts(),labels=['Succeed','Not Succeed'],autopct='%1.2f%%',shadow=True,explode=[0,0.1],colors=['r','b'])
plt.title("Successed and not Successed Terrorist Incidents",size="20")
plt.axis('equal')
plt.show()
```

## Terrorism in Countries and Cities

### For Countries

```python
sns.set(rc={'figure.figsize':(15,10)},font_scale=1)
plt.title("Top 10 Countries have Terrorist Incidents",size=20)
plt.bar(df['country_txt'].value_counts().head(10).index,df['country_txt'].value_counts().head(10),facecolor='r',edgecolor='b')
plt.show()
```

```python
plt.title("Top 10 countries where terrorism occured and its Success",size=20)
sns.countplot(x="country_txt",hue="success",data=df,order=df.country_txt.value_counts().iloc[:10].index)
```

### For Cities

```python
sns.set(rc={'figure.figsize':(20,10)},font_scale=1)
plt.title("Top 10 Cities face Terrorist Incidents",size=20)
plt.bar(df['provstate'].value_counts().head(10).index,df['provstate'].value_counts().head(10),facecolor='r',edgecolor='b')
plt.show()
```

```python
plt.title("Top 10 Cites where terrorism occured and its Success",size=20)
sns.countplot(x="provstate",hue="success",data=df,order=df.provstate.value_counts().iloc[:10].index)
```

## Terrorism via Years

#### Terrorist incidents success from 1970 to 2018

```python
sns.set(rc={'figure.figsize':(20,10)},font_scale=1.4)
plt.plot(df.groupby('iyear').sum()['success'],color='b',ls='-',lw=8,alpha=0.85)
plt.title("Terrorist Incidents via Years",size=20)
plt.xlabel('Years',color='k',size=16) # Add X-axis Label
plt.ylabel('Terrorist Incidents',color='k',size=16)
plt.show()
```

#### Number of persons killed from 1970 to 2018 because of terrorism incidents

```python
sns.set(rc={'figure.figsize':(20,10)},font_scale=1.4)
plt.plot(df.groupby('iyear').sum()['nkill'],color='b',ls='-',lw=8,alpha=0.85)
plt.title("Number of Kills Because of Terrorist Incidents",size=20)
plt.xlabel('Years',color='k',size=16) # Add X-axis Label
plt.ylabel('Number of kills',color='k',size=16)
plt.show()
```
